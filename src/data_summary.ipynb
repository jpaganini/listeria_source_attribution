```python
import pandas as pd

source_1_path='../results/03_human_predictions/Probabilities_Source_1.tsv'
output_path='../results/03_human_predictions/summary'

source_1 = pd.read_csv(source_1_path, sep='\t', header=0)
print(source_1)

source_1_counts = source_1['predictions'].value_counts().sort_values(ascending=False)
print(source_1_counts)

# Calculate the percentage
total_count = source_1_counts.sum()
source_1_counts_with_percentage = pd.DataFrame({'Count': source_1_counts, 'Percentage': (source_1_counts / total_count * 100).round(2)})
print(source_1_counts_with_percentage)

#Export file
source_1_counts_with_percentage.to_csv('../results/03_human_predictions/summary/summary_Source_1.tsv', sep='\t')
```
